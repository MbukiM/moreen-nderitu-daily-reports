### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.

```python
# Exercise 16
```

````python
# Values 
Groups = {
          'GroupA' : np.array([3, 5, 7]),
          'GroupB' : np.array([2, 6, 4]),
         }
categories = ['X', 'Y', 'Z']

# Plot figure
plt.figure(figsize = (8,8))

width = 0.5

# Bottom
bottom = np.zeros(len(categories))

# Plot stacked bar chart using for 
for group_name, group_values in Groups.items():
    plt.bar(categories, group_values, width, label=group_name, bottom=bottom)
    bottom += group_values  # Update bottom to stack the bars

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')

# Add legend
plt.legend()

# Show plot
plt.show()
````

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

```python
# Exercise 20
```

````python
# Cosine and sine Plot 
x = np.linspace(-np.pi , np.pi, 100, endpoint = True)
cos , sin = np.cos(x) , np.sin(x)

# plot figure 
plt.figure(figsize = (10,10),dpi = 100)

# Subplot 
plt.subplot(111)

# plot 
#plt.plot(cos,sin) draws a circle
plt.plot(x, cos, color = "purple" , linewidth = 1.5, linestyle = "dashed",label = "Cos")
plt.plot(x,sin, color = "black", linewidth = 1.6, linestyle = "solid", label = "Sin")

#Labelling
plt.xlabel('x')
plt.ylabel('Cos and sin')
plt.title('Plot for cos and sin')

# Plot axis limit 
plt.xlim(x.min()*1.1, x.max()*1.1)
#plt.ylim(cos.min()*1.1, cos.max()*1.1)
plt.ylim(sin.min()*1.1, sin.max()*1.1)


# Axes Ticks 
plt.xticks([-np.pi, -np.pi/2, 0, +np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([ -1,-0.5,0 , 0.5, 1],[r'$-1$',r'$-0.5$', r'$0$', r'$0.5$', r'$1$'])

# Legend 
plt.legend(loc="upper right", frameon=True)


# optionally, save the figure as a pdf using 72 dots per inch
plt.savefig("./sin_cos.pdf", format='pdf', dpi=80)

#Grid 
plt.grid()
plt.show()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.

```python
# Exercise 21
```

````python
# X and y value 
n = 50
x = np.linspace(-5,5,n)
y = np.linspace(-5,5,n)
x,y = np.meshgrid(x, y)

#Function
def z(x,y): 
    return np.sin(x) + np.cos(y)

# Plot of contour
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.contourf(x, y, z(x,y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(x, y, z(x,y), 8, colors='black')
plt.clabel(C, inline=1, fontsize=10)

plt.xticks([]), plt.yticks([])
plt.show()
````
