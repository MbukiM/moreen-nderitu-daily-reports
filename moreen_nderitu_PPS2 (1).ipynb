### Damped Oscillator

Using odeint, write a code that produces the following two graphs for the simple harmonic
oscillator.

- First graph: position $x(t)$ and velocity $x˙ (t)$ which are both functions of time.

```python
# Importing all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def damped_oscillator(var, p, w_0):
    """
    This function is used to plot Position and Velocity as a function of time
    Input: 
    w_0 = initial angular frequency 
    p = period
    var = displacement and velocity 
    Output:
    Position and Velocity as a function of time
    """
    s, v = var
    dxdt = v
    dvdt = -0.2 * v - w_0**2 * s
    return [dxdt, dvdt]

# Declaring variable
w_0 = 1
s_0 = 1
v_0 = 0
initial_var = [s_0, v_0]
p = np.linspace(0, 20, 400)

#solve the ODE 
solution = odeint(damped_oscillator, initial_var, p, args=(w_0,))

# geting position amd velocity
s = solution[:, 0]
v = solution[:, 1]

# potential energy U(t) and kinetic energy K(t)
U = 0.5 * w_0**2 * s**2
K = 0.5 * v**2

# Plot for position and velocity with respect to time
plt.figure(figsize=(12, 6))
plt.plot(p, s, label='Position s(t)', color='brown', linestyle='-')
plt.plot(p, v, label='Velocity v(t)', color='purple', linestyle='--')
plt.title('Position and Velocity as a function of time')
plt.xlabel('Time (t)')
plt.ylabel('Position and Velocity')
plt.legend()
plt.show()

```

## Explanation 
In a damped oscillator, the position and velocity oscillate with a phase difference of 
π/2, while their amplitudes decay over time due to the damping force. The system eventually loses all its energy, causing both position and velocity to approach zero as the system settles at its equilibrium position.

*(b)*
- Second graph: potential energy $U(t)$ and kinetic energy $K(t)$ which are both functions
of time.

```python
# Plotting potential and kinetic energy as functions of time (Second Graph)
plt.figure(figsize=(12, 6))
plt.plot(p, U, label='Potential Energy U(t)', color='yellow', linestyle='-')
plt.plot(p, K, label='Kinetic Energy K(t)', color='orange', linestyle='--')
plt.title('Potential Energy and Kinetic Energy as a function of time')
plt.xlabel('Time (t)')
plt.ylabel('Potential Energy and Kinetic Energy')
plt.legend()
plt.show()
```

## Explanation
In a damped oscillator, kinetic energy and potential energy continue to oscillate as energy is transferred back and forth between them. However, damping causes both forms of energy to diminish over time, leading to an eventual cessation of motion as the total mechanical energy of the system approaches zero. The key relationship between kinetic and potential energy remains that they oscillate out of phase with each other, but with damping, the total energy of the system gradually decreases.
